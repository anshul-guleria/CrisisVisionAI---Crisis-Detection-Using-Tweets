```python
import os
import pandas as pd

BASE_PATH = "/kaggle/input/crisismmd"
ANNOTATION_PATH = os.path.join(BASE_PATH, "annotations")

all_data = []

for file in os.listdir(ANNOTATION_PATH):
    if file.endswith(".tsv"):
        
        disaster_type = file.replace("_final_data.tsv", "")
        file_path = os.path.join(ANNOTATION_PATH, file)
        
        df = pd.read_csv(file_path, sep="\t")
        df["disaster_type"] = disaster_type
        
        all_data.append(df)

# Combine all events
data = pd.concat(all_data, ignore_index=True)

print("Total samples:", len(data))
print("Columns:", data.columns)
data.head()
```

## Model training

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import pandas as pd
import numpy as np
from PIL import Image

from transformers import AutoTokenizer, AutoModel
from torchvision import models, transforms
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
MAX_LEN = 128
```

```python
image_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
class CrisisDataset(Dataset):
    def __init__(self, df):
        self.df = df.reset_index(drop=True)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # Text
        encoding = tokenizer(
            row["tweet_text"],
            truncation=True,
            padding="max_length",
            max_length=MAX_LEN,
            return_tensors="pt"
        )

        input_ids = encoding["input_ids"].squeeze(0)
        attention_mask = encoding["attention_mask"].squeeze(0)

        # Image
        image = Image.open(row["image_full_path"]).convert("RGB")
        image = image_transform(image)

        label = torch.tensor(row["label_id"], dtype=torch.long)

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "image": image,
            "label": label
        }
```

```python
train_dataset = CrisisDataset(train_df)
val_dataset = CrisisDataset(val_df)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8)
```

```python
class MultiModalModel(nn.Module):
    def __init__(self, num_classes):
        super(MultiModalModel, self).__init__()

        # BERT
        self.bert = AutoModel.from_pretrained("bert-base-uncased")

        # ResNet
        self.cnn = models.resnet50(pretrained=True)
        self.cnn.fc = nn.Identity()  # remove classifier

        # Fusion layers
        self.fc1 = nn.Linear(768 + 2048, 512)
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(512, num_classes)

    def forward(self, input_ids, attention_mask, image):
        # Text
        bert_outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask
        )
        text_features = bert_outputs.last_hidden_state[:, 0, :]  # CLS

        # Image
        image_features = self.cnn(image)

        # Fusion
        combined = torch.cat((text_features, image_features), dim=1)

        x = self.fc1(combined)
        x = torch.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)

        return x
```

```python
model = MultiModalModel(NUM_CLASSES).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=2e-5)
```

```python
from tqdm import tqdm
```

```python
EPOCHS = 3

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}")

    for batch in progress_bar:

        optimizer.zero_grad()

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        images = batch["image"].to(device)
        labels = batch["label"].to(device)

        outputs = model(input_ids, attention_mask, images)

        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        preds = torch.argmax(outputs, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        # Update tqdm bar
        progress_bar.set_postfix({
            "Loss": f"{loss.item():.4f}",
            "Acc": f"{correct/total:.4f}"
        })

    print(f"\nEpoch {epoch+1} Summary:")
    print(f"Avg Loss: {total_loss/len(train_loader):.4f}")
    print(f"Train Accuracy: {correct/total:.4f}")
```

```python
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for batch in val_loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        images = batch["image"].to(device)
        labels = batch["label"].to(device)

        outputs = model(input_ids, attention_mask, images)
        preds = torch.argmax(outputs, dim=1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)

print("Validation Accuracy:", correct / total)
```

```python
import torch
import torch.nn.functional as F
from PIL import Image

def predict_single(text, image_path):

    model.eval()

    # ---- TEXT ----
    encoding = tokenizer(
        text,
        truncation=True,
        padding="max_length",
        max_length=MAX_LEN,
        return_tensors="pt"
    )

    input_ids = encoding["input_ids"].to(device)
    attention_mask = encoding["attention_mask"].to(device)

    # ---- IMAGE ----
    image = Image.open(image_path).convert("RGB")
    image = image_transform(image).unsqueeze(0).to(device)

    # ---- MODEL ----
    with torch.no_grad():
        outputs = model(input_ids, attention_mask, image)

        probs = F.softmax(outputs, dim=1)
        confidence, pred_class = torch.max(probs, dim=1)

    predicted_label = label_encoder.inverse_transform(
        [pred_class.item()]
    )[0]

    return {
        "prediction": predicted_label,
        "confidence": float(confidence.item())
    }
```

```python
from transformers import pipeline
```

```python
import re

def preprocess_for_ner(text):
    # Remove URLs
    text = re.sub(r"http\S+", "", text)

    # Remove mentions
    text = re.sub(r"@\w+", "", text)

    # Split hashtags properly
    text = re.sub(r"#", "", text)

    # Normalize capitalization (important!)
    text = text.title()

    return text.strip()
```

```python
ner_pipeline = pipeline(
    "ner",
    model="Davlan/xlm-roberta-base-ner-hrl",
    aggregation_strategy="simple"
)

def extract_location(text):
    clean_text = preprocess_for_ner(text)
    entities = ner_pipeline(clean_text)
    locations = [
        ent["word"]
        for ent in entities
        if ent["entity_group"] in ["LOC", "GPE"]
    ]

    return locations
```

```python
def predict_with_location(text, image_path):
    
    disaster_result = predict_single(text, image_path)
    
    locations = extract_location(text)
    
    return {
        "disaster_type": disaster_result["prediction"],
        "confidence": disaster_result["confidence"],
        "locations_detected": locations if locations!=[] else "Unknown"
    }
```

```python
random_index = np.random.randint(0, len(val_df))

sample = val_df.iloc[random_index]

result = predict_with_location(
    sample["tweet_text"],
    sample["image_full_path"]
)

print("Tweet:", sample["tweet_text"])
print("True Label:", label_encoder.inverse_transform([sample["label_id"]])[0])
print("Prediction:", result)
```
